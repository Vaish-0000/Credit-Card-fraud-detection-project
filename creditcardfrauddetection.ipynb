```python
# Class distribution: Visualize the distribution of fraud and legit transactions
plt.figure(figsize=(5,4))
sns.countplot(x='Class',data=data)
plt.title('Class Distribution (0: legit, 1: fraud)')
plt.show()
```

```python
# Correlation Matrix: Shows relationships between the features
plt.figure(figsize=(10, 5))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Matrix of the Features')
plt.show()
```

```python
# Boxplot for V1 to analyze its distribution between fraud and non-fraud transactions
plt.figure(figsize=(6, 4))
sns.boxplot(x='Class', y='V1', data=data)
plt.title('Boxplot of V1 - Fraud vs Legit')
plt.show()
```

```python
# Boxplot for V2 to analyze its distribution between fraud and non-fraud transactions
plt.figure(figsize=(6, 4))
sns.boxplot(x='Class', y='V2', data=data)
plt.title('Boxplot of V2 - Fraud vs Legit')
plt.show()
```

```python
# Boxplot for V3 to analyze its distribution between fraud and non-fraud transactions
plt.figure(figsize=(6, 4))
sns.boxplot(x='Class', y='V3', data=data)
plt.title('Boxplot of V3 - Fraud vs Legit')
plt.show()
```

```python
# Focus on correlations with the target variable (Class)
plt.figure(figsize=(10,5))
correlation_with_target = correlation_matrix['Class'].sort_values(ascending=False)
sns.barplot(x=correlation_with_target.index, y=correlation_with_target.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Correlation of Features with Fraud Class')
plt.show()
```
